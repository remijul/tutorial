## Introduction
Data analysis on diamonds dataset ([more info here](https://ggplot2.tidyverse.org/reference/diamonds.html)).

## Data exploration

```
# URL for collecting dataset
url_csv = "https://fxjollois.github.io/cours-2016-2017/donnees/diamonds.csv"
```

```
# Import the Pandas library
import pandas as pd
```

```
# Read the dataset with the .read_csv() method
df = pd.read_csv(url_csv) 
```

```
# Some information about the dataset
df.info()
```

```
# View the head of the dataset
df.head()
```

```
# View the tail of the dataset
df.tail()
```

```
# List of columns
list(df.columns)
```

```
# Check of presence of NA's in the dataset
df.isna().sum()
```

## Univariate analysis

### Statistical description

```
# Statistical description with .describe() method : for quantitative data only
df.describe()
```

```
# Statistical description with .describe() method : for qualitative AND quantitative data only
df.describe(include='all')
```

```
# Compute mean for all variables
df.mean()
```

```
# Compute other statistics for all variables
df.median()
#df.min()
#df.max()
```

```
# Compute quantiles for price
df.price.quantile([0.01, 0.25, 0.5, 0.75, 0.99])
```

### Graphical description

```
# Plot histogram of all quantitative variables : not really usefull here (too much variables)
df.plot.hist()
```

```
# Plot histogram of price variable
df.price.hist(bins = 20)
```

```
# Plot histogram of price variable
df.boxplot(column='price')
```

## Bivariate analysis

```
# Correlation matrix between all numerical variables
df.corr(method='pearson')
```

```
# Plot heatmap with Pandas
df.corr().style.background_gradient(cmap='coolwarm', axis=None).set_precision(2).set_properties(**{'font-size': '8pt'})
```

```
# Plot heatmap with seaborn
import seaborn as sns
sns.heatmap(df.corr(), annot=True, linewidths=0.5)
```

```
# Plot pair variable analysis
sns.pairplot(df, hue='cut')
```
