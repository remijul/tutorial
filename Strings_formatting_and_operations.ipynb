# Strings formatting and operations
In computer programming, a **string** is traditionally a **sequence of characters**, either as a literal constant or as some kind of variable. The latter may allow its elements to be mutated and the length changed, or it may be fixed (after creation). A string is generally considered as a data type and is often implemented as an **array data structure of bytes** (or words) that stores a sequence of elements, typically characters, using some **character encoding**.  
Source : [Wikipedia](https://en.wikipedia.org/wiki/String_(computer_science))

## Strings formatting
This concept is particularly useful when you want to create more complex strings and messages.  
Use the "%" operator to build the final message.

```
string1 = "Python"
print("I love %s!" % string1)
```

For several elements simply use the tuple structure, with parentheses, and create your own formatting.

```
string2 = "beginner"
print("I'm a %s and I love %s!" % (string2, string1))
```

## Strings operations
Many operations on strings are available with Python, the most frequent are described below.

###Concatenating 
The `+` operator is simply use to concatenate different strings to make a longer one.

```
myname = "Yoda"
print("Hello " + myname + ", fun you have ?")
```

The `*` operator is simply use to multiply strings to make a longer one.

```
myname = "Yoda"
print("When I see " + 10 * myname + ", I feel good !")
```

### Indexing
Each character in a string get an index, so indexing a string is frequently needed.  
In Python the first element / index get the **0 position**.  
**Remember the 0 position you should !**

```
myname[0]
```

```
myname[3]
```

```
myname[4]
```

Counting from the right is possible with the `-` operator.

```
myname[-4]
```

###Slicing
Slicing is a way to extract different elements of the strings given a standard method : `mystring[start:end:step]`
- start : the index for starting the slicing, index is **included**
- end : the index for ending the slicing, index is **excluded**
- step : the step between each index for slicing, **1 is the default**


The complete string since the step is set to 1.

```
myname[0:4:1]
```

One out two characters.

```
myname[0:4:2]
```

The **end** index is excluded.

```
myname[0:3:1]
```

This syntax is more concise but still correct and gives the same results.

```
myname[:3]
```

Other example.

```
10*myname[0:3:1]
```

### The built-in methods for strings
The string type allows to use lot of **built-in** methods.  
A built-in method in Python is a already implemented program for performing a specific task.

#### Length of string
How many characters in a string ?  The `len` function is the way !

```
len(myname)
```

```
len(10*myname)
```

####Index of character
Let's imagine you want to retrieve the index of specific character into the string. The `.index()` method is the way !

```
myname.index("o")
```

```
myname.index("Y")
```

Python is very sensitive ! And particulary to the case ...

```
myname.index("y")
```

Only the first matched occurence is given back !

```
longstring = "May the Force be with you"
longstring.index("y")
```

#### Count of character
Let's imagine you want to count how many characters are present into the string. The .count() method is the way !

```
longstring.count("y")
```

#### Change the case
Sometimes operations for changing the case of the string are needed. Both `.lower()` and `.upper()` methods are the way ! 

```
longstring.lower()
```

```
longstring.upper()
```

#### Split the string
Let's imagine you want to divide the string into several pieces considering a specific character as pattern of division. The .split() method is the way !

```
longstring.split()
```

The default pattern of division is the blank space but you can specify your own.

```
longstring.split("e")
```

```
longstring.split("y")
```

####Match start / end of the string
Let's imagine you want to check if the string starts or ends with a given pattern of string. Both `.startswith()` and `.endswith()` methods are the way !

```
longstring.startswith("May")
```

```
longstring.startswith("the dark side")
```

```
longstring.endswith("you")
```

## Exercices

The goal of this exercice is to find the correct sentence `s` corresponding to the statements give in the code.

```
s = "Much to learn, you still have."

# Length should be 30
print("Length of s = %d" % len(s))

# First occurrence of "l" should be at index 8
print("The first occurrence of the letter l = %d" % s.index("l"))

# Number of l's should be 3
print("l occurs %d times" % s.count("l"))

# Slicing the string into bits
print("The first five characters are '%s'" % s[:5]) # Start to 5
print("The next five characters are '%s'" % s[5:10]) # 5 to 10
print("The thirteenth character is '%s'" % s[12]) # Just number 12
print("The characters with odd index are '%s'" %s[1::2]) #(0-based indexing)
print("The last five characters are '%s'" % s[-5:]) # 5th-from-last to end

# Convert everything to uppercase
print("String in uppercase: %s" % s.upper())

# Convert everything to lowercase
print("String in lowercase: %s" % s.lower())

# Check how a string starts
if s.startswith("Much to"):
    print("String starts with 'Much to'. Good!")

# Check how a string ends
if s.endswith("you still have."):
    print("String ends with 'you still have.'. Good!")

# Split the string into 6 separate strings,
# each containing only a word
print("Split the words of the string: %s" % s.split(" "))
```
